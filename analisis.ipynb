```python
%matplotlib inline
import pandas as pd
import seaborn as sns
```

```python
musimundo = pd.read_json('https://raw.githubusercontent.com/ciberlunes/ciberlunes.github.io/master/jsons/hs_mm_17.json')
musimundo = musimundo.rename(columns={'pr': 'Descuento Real', 'pv': 'Descuento promocionado', 'or': 'Precio anterior', 'cm': 'Precio con promoción', 's': 'Supuesto precio anterior'})
musimundo
```

```python
sns.pairplot(musimundo)
```

```python
ax = sns.pairplot(musimundo, vars=['Descuento Real', 'Descuento promocionado'], hue='Precio anterior', size=5)
ax
```
