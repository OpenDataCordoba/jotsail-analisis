```python
%matplotlib inline
import pandas as pd
import seaborn as sns
```

```python
def get_and_rename(json_dataset):
    ds = pd.read_json(json_dataset)
    return ds.rename(columns={'pr': 'Descuento Real', 'pv': 'Descuento promocionado', 'or': 'Precio anterior', 'cm': 'Precio con promoci칩n', 's': 'Supuesto precio anterior'})
```

## Musimundo

```python
musimundo = get_and_rename('https://raw.githubusercontent.com/ciberlunes/ciberlunes.github.io/master/jsons/hs_mm_17.json')
```

```python
sns.pairplot(musimundo)
```

```python
ax = sns.pairplot (musimundo, vars=['Descuento Real', 'Descuento promocionado'], 
              hue='Precio anterior', size=5, palette='viridis_r')
```
