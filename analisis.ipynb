```python
%matplotlib inline
import pandas as pd
import seaborn as sns
```

```python
def get_and_rename(json_dataset):
    ds = pd.read_json(json_dataset)
    return ds.rename(columns={'pr': 'Descuento Real', 'pv': 'Descuento promocionado', 'or': 'Precio anterior', 'cm': 'Precio con promoci칩n', 's': 'Supuesto precio anterior'})
    
```

## Musimundo

```python
musimundo = get_and_rename('https://raw.githubusercontent.com/ciberlunes/ciberlunes.github.io/master/jsons/hs_mm_17.json')
fravega = get_and_rename('https://github.com/ciberlunes/ciberlunes.github.io/raw/master/jsons/hs_fv_17.json')
compumundo = get_and_rename('https://raw.githubusercontent.com/ciberlunes/ciberlunes.github.io/master/jsons/hs_cpm_17.json')
garbarino = get_and_rename('https://github.com/ciberlunes/ciberlunes.github.io/raw/master/jsons/hs_gb_17.json')
```

```python
ax = sns.distplot(musimundo['Descuento Real'])
ax.legend = 'real'

ax = sns.distplot(musimundo['Descuento promocionado'], axlabel='descuento real', color='red', ax=ax)
ax.legend = 'promocionado'
ax.set_title('Distribucion de descuentos real (azul) vs promocionado (rojo) en Musimundo')
ax.set_xlabel('')
```

```python
sns.pairplot(musimundo)
```

```python
ax = sns.pairplot(musimundo, vars=['Descuento promocionado', 'Descuento Real'], 
                  hue='Precio anterior', size=5, palette='viridis_r', grid_kws={'despine': True})
```
